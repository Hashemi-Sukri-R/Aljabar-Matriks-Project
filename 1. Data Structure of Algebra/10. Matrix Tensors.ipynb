```python
import numpy as np
```

```python
# buat matrix X berukuran 2x3 menggunakan numpy
X = np.array([[10, 20, 30], [40, 50, 60]])
print(X)
```

```python
# tampilkan ukuran dari matrix X
print(f"X: {X.shape}")
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
print(X[:, 0])
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X
print(X[:, 1])
```

```python
import torch as pt
```

```python
# buat matrix X berukuran 2x3
X = pt.tensor([[10, 20, 30], [40, 50, 60]])
print(X)
```

```python
# tampilkan ukuran dari matrix X
print(f"X: {X.shape}")
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
print(X[:, 0].tolist())
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
print(X[:, 1].tolist())
```

```python
# ulangi soal nomor 1 menggunakan tensorflow
```

```python
import tensorflow as tf
```

```python
# buat matrix X berukuran 2x3
X = tf.Variable([[10, 20, 30], [40, 50, 60]])
print(X)
```

```python
# tampilkan ukuran dari matrix X
print(f"X: {X.shape}")
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
print(X[:, 0].numpy())
```

```python
# tampilkan seluruh elemen dari kolom kedua dari matrix X
print(X[:, 1].numpy())
```
