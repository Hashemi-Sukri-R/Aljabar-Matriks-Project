1. SKALAR

```python
# buat variabel skalar x dan simpan suatu nilai skalar
x = 8
print(x)
```

```python
# tampilkan tipe data dari x
type(x)
```

```python
# buat variabel skalar y dan simpan suatu nilai skalar
y = 10
print(y)
```

```python
# tampilkan tipe data dari y
type(y)
```

```python
# jumlahkan x + y
py_sum = x + y
print(py_sum)
```

```python
# buat variabel x_float (nilai desimal)
x_float = 2.5
print(x_float)
```

```python
# jumlahkan x_float dengan y
x_float = 2.5
float_sum = x_float + y
print(float_sum)
```

```python
# tampilkan tipe data x_float
type(float_sum)
```

2. TENSORFLOW

```python
# lakukan tahapan pada soal 1 dengan menggunakan tensorflow
import tensorflow as tf
```

```python
# buat variabel tensorflow x dan simpan nilai tensorflow serta tipe datanya
x_tf = tf.Variable(8, dtype = tf.int32)
print(x_tf)
```

```python
# buat variabel tensorflow y dan simpan nilai tensorflow serta tipe datanya
y_tf = tf.Variable(10, dtype = tf.int32)
print(y_tf)
```

```python
# jumlahkan x_tf + y_tf
tf_sum = tf.add(x_tf, y_tf)
print(tf_sum)
```

```python
# buat variabel bernama x_float_tf dan simpan suatu nilai desimal serta tipe datanya
x_float_tf = tf.Variable(2.5, dtype = tf.float32)
print(x_float_tf)
```

```python
# jumlahkan x_float_tf dengan y_tf
y_float_tf = float(y_tf)
penjumlahan = x_float_tf + y_float_tf
print(penjumlahan)
```

3. PYTORCH

```python
# lakukan tahapan pada soal 1 dengan menggunakan pytorch
import torch as pt
```

```python
# buat variabel pytorch x dan simpan nilai pytorch serta tipe datanya
x_pt = pt.tensor(8, dtype = pt.int32)
print(x_pt)
```

```python
# buat variabel pytorch y dan simpan nilai pytorch serta tipe datanya
y_pt = pt.tensor(10, dtype = pt.int32)
print(y_pt)
```

```python
# jumlahkan x_pt + y_pt
pt_sum = pt.add(x_pt, y_pt)
print(pt_sum)
```

```python
# buat variabel bernama x_float_pt dan simpan suatu nilai desimal serta tipe datanya
x_float_pt = pt.tensor(2.5, dtype = pt.float32)
print(x_float_pt)
```

```python
# jumlahkan x_float_pt dengan y_pt
y_float_pt = float(y_pt)
penjumlahan = x_float_pt + y_float_pt
print(penjumlahan)
```
