PYTORCH

```python
import torch as pt
```

```python
x_pt = pt.tensor([12, 8, 25])
print(x_pt)
x_pt.shape
```

```python
x_pT = x_pt.T
print(x_pT)
x_pT.shape
```

```python
y_pt = pt.tensor([7, 79, 77])
print(y_pt)
y_pt.shape
```

```python
y_pT = y_pt.T
print(y_pT)
y_pT.shape
```

TENSORFLOW

```python
import tensorflow as tf
```

```python
x_tf = tf.Variable([12, 8, 25], dtype = tf.int32)
print(x_tf)
x_tf.shape
```

```python
x_tF = tf.transpose(x_tf)
print(x_tF)
x_tF.shape
```

```python
y_tf = tf.Variable([7, 79, 77], dtype = tf.int32)
print(y_tf)
y_tf.shape
```

```python
y_tF = tf.transpose(y_tf)
print(y_tF)
y_tF.shape
```
