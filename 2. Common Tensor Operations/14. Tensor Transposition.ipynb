```python
import numpy as np
import torch
import tensorflow as tf
```

```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array([[50, 67, 32], [27, 67, 45]])
print(X)
```

```python
# buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[50, 67, 32], [27, 67, 45]])
print(X_pt)
```

```python
# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable([[50, 67, 32], [27, 67, 45]])
print(X_tf)
```
