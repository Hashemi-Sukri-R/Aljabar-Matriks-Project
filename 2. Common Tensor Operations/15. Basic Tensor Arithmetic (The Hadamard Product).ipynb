# Basic Tensor Arithmetic  (The Hadamard Product)

No. 1

```python
import numpy as np
import torch
import tensorflow as tf
```

```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array([[34, 56, 23], [12, 35, 63]])
X
```

```python
# buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[40, 12, 53], [23, 87, 34]])
X_pt
```

```python
# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable([[4, 3, 2], [5, 6, 0]])
X_tf
```

**Perkalian matrix X, X_pt dan X_tf dengan suatu skalar**

```python
X1 = X * 5
X1
```

```python
X2 = X_pt * 5
X2
```

```python
X3 = X_tf * 5
X3
```

**penjumlahan matrix X, X_pt dan X_tf dengan suatu skalar**

```python
Xa = X + 10
Xa
```

```python
Xb = X_pt + 10
Xb
```

```python
Xc = X_tf + 10
Xc
```

**Pengurangan matrix X, X_pt dan X_tf dengan suatu skalar**

```python
XA = X - 12
XA
```

```python
XB = X_pt - 12
XB
```

```python
XC = X_tf - 12
XC
```

No. 2

```python
# buat matrix Y dengan ukuran 2x3 menggunakan numpy
Y = np.array([[2, 4, 8], [1, 0, 2]])
Y
```

```python
# buat matrix Y_pt dengan ukuran 2x3 menggunakan pytorch
Y_pt = torch.tensor([[4, 7, 9], [4, 7, 2]])
Y_pt
```

```python
# buat matrix Y_tf dengan ukuran 2x3 menggunakan tensorflow
Y_tf = tf.Variable([[3, 6, 9], [0, 2, 7]])
Y_tf
```

```python
# Perkalian matrix X dengan Y
xy = X * Y
xy
```

```python
# Perkalian matrix X_pt dengan Y_pt
XpYp = X_pt * Y_pt
XpYp
```

```python
# Perkalian matrix X_tf dengan Y_tf
XtYt = X_tf * Y_tf
XtYt
```
