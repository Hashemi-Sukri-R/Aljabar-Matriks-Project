```python
import numpy as np
import torch
import tensorflow as tf
```

```python
# buat vector x dan y dengan ukuran 1x3 menggunakan numpy
x = np.array([6, 7, 2])
y = np.array([9, 5, 1])
print("vector x:", x)
print("vector y:", y)
```

```python
# buat vector x_pt dan y_pt dengan ukuran 1x3 menggunakan pytorch
x_pt = torch.tensor([6, 7, 2])
y_pt = torch.tensor([9, 5, 1])
print("vector x_pt:", x_pt)
print("vector y_pt:", y_pt)
```

```python
# buat vector x_tf dan y_tf dengan ukuran 1x3 menggunakan tensorflow
x_tf = tf.Variable([6, 7, 2])
y_tf = tf.Variable([9, 5, 1])
print("vector x_tf:", x_tf)
print("vector y_tf:", y_tf)
```

```python
# hitung perkalian dot product antara vector x dan vector y
npdot = np.dot(x, y)
print("Hasil dot product:", npdot)
```

```python
# hitung perkalian dot product antara vector x_pt dan vector y_pt
ptdot = torch.dot(x_pt, y_pt)
print("Hasil dot product:", ptdot)
```

```python
# hitung perkalian dot product antara vector x_tf dan vector y_tf
tf.reduce_sum(tf.multiply(x_tf, y_tf))
```
