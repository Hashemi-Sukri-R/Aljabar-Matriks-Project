```python
import numpy as np
import matplotlib.pyplot as plt
```

**Buat visualisasi dari persamaan berikut**

$$y = 2x+1$$
$$3x-y = 5$$

Untuk: $$3x - y = 5$$ bisa ditulis seperti:
$$y = 3x - 5$$




```python
x = np.linspace(-10, 10, 1000)
```

```python
y1 = 2*x + 1
y2 = 3*x - 5
```

```python
fig, ax = plt.subplots()
plt.xlabel("x")
plt.ylabel("y")

ax.set_xlim([-1, 15])
ax.set_ylim([-1, 15])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')
plt.axvline(x=6, color='purple', linestyle='--')
plt.axhline(y=13, color='purple', linestyle='--')
```
