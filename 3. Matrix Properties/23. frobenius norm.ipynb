SOAL NO 1. menggunakan numpy, pytorch dan tensorflow buatlah matrix ukuran 3x3 lalu hitung nilai frobenius norm

1. NUMPY

```
import numpy as np
```

```
X = np.array([[4, 1, 2], [7, 3, 5], [9, 8, 6]])
X
```

```
(4**2 + 1**2 + 2**2 + 7**2 + 3**2 + 5**2 + 9**2 + 8**2 + 6**2)**(1/2)
```

```
np.linalg.norm(X) #L2 norm
```

2. PYTORCH

```
import torch
```

```
X_pt = torch.tensor([[4, 1, 2], [7, 3, 5], [9, 8, 6.]]) # torch.norm() supports floats only
X_pt
```

```
torch.norm(X_pt)
```

3. TENSORFLOW

```
import tensorflow as tf
```

```
X_tf = tf.Variable([[4, 1, 2], [7, 3, 5], [9, 8, 6.]]) # tf.norm() also supports floats only
X_tf
```

```
tf.norm(X_tf)
```
